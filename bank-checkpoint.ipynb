```python
Y=df['y']
```

```python
df['job'] = df['job'].astype('category').cat.codes
df['marital'] = df['marital'].astype('category').cat.codes
df['education'] = df['education'].astype('category').cat.codes
df['contact'] = df['contact'].astype('category').cat.codes
df['poutcome'] = df['poutcome'].astype('category').cat.codes
df['month'] = df['month'].astype('category').cat.codes
df['default'] = df['default'].astype('category').cat.codes
df['loan'] = df['loan'].astype('category').cat.codes
df['housing'] = df['housing'].astype('category').cat.codes
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(df.drop(['day_of_week','y'], axis=1),Y,test_size=0.2,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
lc=DecisionTreeClassifier()
lc.fit(X_train,Y_train)
```

```python
lc.feature_importances_
```

```python
importances = lc.feature_importances_
feature_names = df.drop(['y','day_of_week'], axis=1).columns
indices = np.argsort(importances)

def feature_importance_graph(indices, importances, feature_names):
    plt.figure(figsize=(12,6))
    plt.title("Feature Importance", fontsize=10)
    plt.barh(range(len(indices)), importances[indices], color='r',  align="center")
    plt.yticks(range(len(indices)), feature_names[indices], rotation='horizontal',fontsize=14)
    plt.ylim([-1, len(indices)])
    
feature_importance_graph(indices, importances, feature_names)
plt.show()
```

```python
X1=df[['duration','age','euribor3m','campaign','job','education','marital','month','housing','loan']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X1,Y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier()
rc.fit(x_train,y_train)
```

```python
pred=rc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
a1=accuracy_score(y_test,pred)
a1
```

```python
confusion_matrix(y_test,pred)
```

```python
from xgboost import XGBClassifier
xg=XGBClassifier()
xg.fit(x_train,y_train)
pred1=xg.predict(x_test)
a2=accuracy_score(y_test,pred1)
a2
```

```python
confusion_matrix(y_test,pred1)
```

```python
score=[a1,a2]
col=['score']
model=['random forest','xgboost']
data=pd.DataFrame(data=score,index=model,columns=col)
data
```
