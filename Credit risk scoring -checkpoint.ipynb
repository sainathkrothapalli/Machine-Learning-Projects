```python
df.columns = ['chk_acct','duration','credit_his','purpose','amount','saving_acct',
               'present_emp','installment_rate','sex','other_debtor','present_resid','property',
               'age','other_install','housing','n_credits','job','n_people','telephone','foreign','response']
df.head()
```

```python
x= df[['age', 'sex', 'housing', 'saving_acct',
       'chk_acct', 'duration','amount']]
Y= df['response']
```

```python
x.dtypes
```

```python
cat_cols = list(x.columns[x.dtypes == 'object'])
num_cols = list(x.columns[x.dtypes != 'object'])
x_dummy =  pd.get_dummies(x[cat_cols], drop_first=True)
x_new = pd.concat([x_dummy, x[num_cols]], axis=1, join='inner')
```

```python
from sklearn.preprocessing import StandardScaler
s= StandardScaler()
X= s.fit_transform(x_new)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier()
rc.fit(X_train,Y_train)
```

```python
y_pred=rc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(Y_test,y_pred)
cm
```

```python
a1=accuracy_score(Y_test,y_pred)
a1
```

```python
from sklearn.svm import SVC
sc=SVC()
sc.fit(X_train,Y_train)
y_pred1=sc.predict(X_test)
```

```python
cm=confusion_matrix(Y_test,y_pred1)
cm
```

```python
a2=accuracy_score(Y_test,y_pred1)
a2
```

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier()
dc.fit(X_train,Y_train)
y_pred2=dc.predict(X_test)
```

```python
cm=confusion_matrix(Y_test,y_pred2)
cm
```

```python
a3=accuracy_score(Y_test,y_pred2)
a3
```

```python
from sklearn.model_selection import cross_val_score
cvs=cross_val_score(rc,X_train,Y_train,scoring='accuracy',cv=5)
cvs
```

```python
cvs.mean()
```

```python
from xgboost import XGBClassifier
xg=XGBClassifier(n_estimators=500)
xg.fit(X_train,Y_train)
y_pred3=xg.predict(X_test)
```

```python
cm=confusion_matrix(Y_test,y_pred3)
cm
```

```python
a4=accuracy_score(Y_test,y_pred3)
a4
```
